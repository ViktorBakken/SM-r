```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
df= pd.read_csv(os.path.abspath("../data/labeled_training_data.csv"))
```

```python
temp_trend  = df.groupby('temp')['increase_stock'].mean()

#Calculate a moving average with a window size of 5
temp_trend_smoothed = temp_trend.rolling(window=5).mean()

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(temp_trend.index, temp_trend.values, label='Original', color='skyblue', alpha=0.5)
plt.plot(temp_trend_smoothed.index, temp_trend_smoothed.values, label='Smoothed', color='navy', linewidth=2)
plt.xlabel("Temperature (°C)")
plt.xticks(ticks=np.arange(0,35,5))
plt.ylabel("Proportion Needing Increase")
plt.title("Proportion of Instances Requiring Bicycle Increase by Temperature")
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
sns.histplot(df['temp'], bins=64, kde=True, color='coral')
plt.xlabel('Temperature')
plt.xticks(ticks=np.arange(0,35,5))
plt.ylabel('Frequency')
plt.title('Distribution of Temperature Values')
plt.show()

```

```python
#Analysis of new features 

# note that('snow', 'snowdepth', 'holiday', 'visibility', 'precip') are obviously bad and should be dropped
```

```python
#Day or night feature. Som man ser på grafen är det nästan aldrig en ökning under natten (bra feature)

#Enl truls graf borde dessa vara de optimala timmarna att välja
df['day_or_night'] = df['hour_of_day'].apply(lambda x: 1 if 8 <= x < 21 else 0)


day_or_night_trend = df.groupby('day_or_night')['increase_stock'].mean()

plt.figure()
day_or_night_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase Depending on Day or not")
plt.xlabel("Not Day / Day")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
#Optimal wind 5 < x <= 31 (kanske är bra)

df['opt_wind'] = df['windspeed'].apply(lambda x: 1 if 5 < x <= 31 else 0)


opt_wind_trend = df.groupby('opt_wind')['increase_stock'].mean()

plt.figure()
opt_wind_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase Depending on Optimal wind or not")
plt.xlabel("Not optimal wind / Optimal Wind")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
#Optimal wind 5 < x <= 31 (bra feature)

df['cold'] = df['temp'].apply(lambda x: 1 if x <= 8 else 0)


cold_trend = df.groupby('cold')['increase_stock'].mean()

plt.figure()
cold_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase Depending on cold or not")
plt.xlabel("Not cold / cold")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
#winter (kanske är bra)

df['winter'] = df['month'].apply(lambda x: 1 if x == 0 or x== 1  or x==11 else 0)


winter_trend = df.groupby('winter')['increase_stock'].mean()

plt.figure()
winter_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase Depending on Winter or not")
plt.xlabel("Not winter / Winter")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')
```
