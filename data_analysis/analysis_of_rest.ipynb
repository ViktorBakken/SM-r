```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
df= pd.read_csv(os.path.abspath("../data/labeled_training_data.csv"))
```

```python
temp_trend  = df.groupby('temp')['increase_stock'].mean()

#Calculate a moving average with a window size of 5
temp_trend_smoothed = temp_trend.rolling(window=5).mean()

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(temp_trend.index, temp_trend.values, label='Original', color='skyblue', alpha=0.5)
plt.plot(temp_trend_smoothed.index, temp_trend_smoothed.values, label='Smoothed', color='navy', linewidth=2)
plt.xlabel("Temperature (C)")
plt.xticks(ticks=np.arange(0,35,5))
plt.ylabel("Proportion Needing Increase")
plt.title("Proportion of Instances Requiring Bicycle Increase by Temperature")
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
sns.histplot(df['temp'], bins=64, kde=True, color='coral')
plt.xlabel('Temperature')
plt.xticks(ticks=np.arange(0,35,5))
plt.ylabel('Frequency')
plt.title('Distribution of Temperature Values')
plt.show()

```

```python
#Day or night feature. Som man ser p grafen r det nstan aldrig en kning under natten (bra feature)

#Enl truls graf borde dessa vara de optimala timmarna att vlja
df['day_or_night'] = df['hour_of_day'].apply(lambda x: 1 if 8 <= x < 21 else 0)


day_or_night_trend = df.groupby('day_or_night')['increase_stock'].mean()

plt.figure()
day_or_night_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase Depending on Day or not")
plt.xlabel("Not Day / Day")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
#Optimal wind 5 < x <= 31 (kanske r bra)

df['opt_wind'] = df['windspeed'].apply(lambda x: 1 if 5 < x <= 31 else 0)


opt_wind_trend = df.groupby('opt_wind')['increase_stock'].mean()

plt.figure()
opt_wind_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase Depending on Optimal wind or not")
plt.xlabel("Not optimal wind / Optimal Wind")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
#winter (kanske r bra)

df['winter'] = df['month'].apply(lambda x: 1 if x == 2 or x== 1  or x==12 else 0)


winter_trend = df.groupby('winter')['increase_stock'].mean()

plt.figure()
winter_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase Depending on Winter or not")
plt.xlabel("Not winter / Winter")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
df['atemp'] = (243.04 * (np.log(df['humidity']/100) + (17.625 * df['dew']) / (243.04 + df['dew']))) / (17.625 - np.log(df['humidity']/100) - (17.625 * df['dew']) / (243.04 + df['dew']))

#df['atemp'] = df['temp'] + 0.33 * df['humidity'] - 0.70 * df['windspeed'] - 4.00


atemp_trend  = df.groupby('atemp')['increase_stock'].mean()

atemp_trend_smoothed = atemp_trend.rolling(window=5).mean()

plt.figure()
plt.plot(atemp_trend.index, atemp_trend.values, label='Original', color='skyblue', alpha=0.5)
plt.plot(atemp_trend_smoothed.index, atemp_trend_smoothed.values, label='Smoothed', color='navy', linewidth=2)
plt.xlabel("Apparent Temprature")
plt.ylabel("Proportion Needing Increase")
plt.title("Proportion of Instances Requiring Bicycle Increase by Apparent temp")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['heat_index'] = (-42.379 + 2.04901523*(df['temp']*9/5 + 32) + 10.14333127*df['humidity'] - 0.22475541*(df['temp']*9/5 + 32)*df['humidity'] - 0.00683783*(df['temp']*9/5 + 32)**2 - 0.05481717*(df['humidity']**2) + 0.00122874*(df['temp']*9/5 + 32)**2*df['humidity'] + 0.00085282*(df['temp']*9/5 + 32)*df['humidity']**2 - 0.00000199*(df['temp']*9/5 + 32)**2*df['humidity']**2) * 5/9


heat_index_trend  = df.groupby('heat_index')['increase_stock'].mean()

heat_index_trend_smoothed = heat_index_trend.rolling(window=10).mean()

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(heat_index_trend.index, heat_index_trend.values, label='Original', color='skyblue', alpha=0.5)
plt.plot(heat_index_trend_smoothed.index, heat_index_trend_smoothed.values, label='Smoothed', color='navy', linewidth=2)
plt.xlabel("Heat Index")
plt.ylabel("Proportion Needing Increase")
plt.title("Proportion of Instances Requiring Bicycle Increase by Heat Index")
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
sns.histplot(df['heat_index'], bins=64, kde=True, color='coral')
plt.xlabel('Cloudcover')
plt.ylabel('Frequency')
plt.title('Distribution of Heat Index Values')
plt.show()
```
