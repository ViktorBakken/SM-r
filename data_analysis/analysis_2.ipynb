# Is there any difference between weekdays and holidays?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Load data
df = pd.read_csv(os.path.abspath("../data/normalized_labeled_training_data.csv"))
df_not_normalized = pd.read_csv(os.path.abspath("../data/labeled_training_data.csv"))
```

```python
weekday_mean = df_not_normalized.groupby('weekday')['increase_stock'].mean()
holiday_mean = df_not_normalized.groupby('holiday')['increase_stock'].mean()
```

```python
# Subplot 1: Weekday Trend
plt.figure(figsize=(6, 4))
weekday_mean.plot(kind='bar', color='skyblue')
plt.title("Proportion of Bicycle Increase: Weekday vs Weekend")
plt.xlabel("0=Weekend, 1=Weekday")
plt.ylabel("Proportion Needing Increase")
plt.xticks(ticks=[0, 1])
plt.grid(axis='y')
plt.show()

# Subplot 2: Holiday Trend
plt.figure(figsize=(6, 4))
holiday_mean.plot(kind='bar', color='orange')
plt.title("Proportion of Bicycle Increase: Holiday vs Not Holiday")
plt.xlabel("0=Not Holiday, 1=Holiday")
plt.ylabel("Proportion Needing Increase")
plt.xticks(ticks=[0, 1])
plt.grid(axis='y')
plt.show()
```

```python
# Subplot 3:

barWidth = 1
bars_weekday = [weekday_mean[0], weekday_mean[1]]
bars_holiday = [holiday_mean[0], holiday_mean[1]]

weekday_positions = np.arange(len(bars_weekday)) 
holiday_positions = weekday_positions + len(bars_weekday) + barWidth 

fig, weekday_holiday = plt.subplots(figsize=(8, 6))
weekday_holiday.bar(weekday_positions, bars_weekday, color=['skyblue', 'orange'])
weekday_holiday.bar(holiday_positions, bars_holiday, color=['lightgreen', 'purple'])

positions = np.concatenate([weekday_positions, holiday_positions]) 
weekday_holiday.set_xticks(positions)
weekday_holiday.set_title("Comparison of Increase Stock: Weekday vs Holiday")
weekday_holiday.set_ylabel("Proportion Needing Increase")
weekday_holiday.set_xticklabels(['Not Weekday', 'Weekday', 'Not Holiday', 'Holiday'], rotation=0)
weekday_holiday.grid(axis='y')
plt.show()
```
