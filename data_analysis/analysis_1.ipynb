# Can any trend be seen comparing different hours, weeks, and months?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#Load data
df = pd.read_csv(os.path.abspath("../data/normalized_labeled_training_data.csv"))
df_not_normalized = pd.read_csv(os.path.abspath("../data/labeled_training_data.csv"))
```

```python
hourly_trend  = df_not_normalized.groupby('hour_of_day')['increase_stock'].mean()
weekly_trend  = df_not_normalized.groupby('day_of_week')['increase_stock'].mean()
monthly_trend = df_not_normalized.groupby('month')['increase_stock'].mean()

# Subplot 1: Hourly Trend
plt.figure(figsize=(12, 12))
plt.subplot(3, 1, 1)
hourly_trend.plot(kind='bar', color='skyblue')
plt.title("Proportion of Instances Requiring Bicycle Increase by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')

# Subplot 2: Weekly Trend
plt.subplot(3, 1, 2)
weekly_trend.plot(kind='bar', color='orange')
plt.title("Proportion of Instances Requiring Bicycle Increase by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Proportion Needing Increase")
plt.xticks(rotation=45)
plt.grid(axis='y')

# Subplot 3: Monthly Trend
plt.subplot(3, 1, 3)
monthly_trend.plot(kind='bar', color='green')
plt.title("Proportion of Instances Requiring Bicycle Increase by Month")
plt.xlabel("Month")
plt.ylabel("Proportion Needing Increase")
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(axis='y')

# Adjust layout and show the plot
plt.tight_layout()
downloads_path = os.path.expanduser("~/Downloads/weekly_trend_plot.png")
plt.savefig(downloads_path, dpi=300, bbox_inches='tight')  # Save as PNG
plt.show()
```
